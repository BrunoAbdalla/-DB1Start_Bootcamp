```python
#6) Aprimore o programa de gerenciamento do jogador (Exercício 04) adicionando vários jogadores e mostre um resumo e permitar
#   ver detalhes de cada jogador:

#4) Crie um programa que gerencie o aproveitamento de um jogador de futebol, o programa vai ler o nome do jogador e qual o 
#   número de partidas que foram feitas e quantos gols ele fez em cada jogo (lista), e tudo isso seja armazenado em um 
#   dicionário, acrescentando também o total de gols.

#   O programa deve retornar: Os dados do dicionário;
#   Os dados do dicionário formatados(ex.: "O campo K tem valor V")
#   O número de gols marcados em cada jogo formatado (ex.: No jogo  n, o jogador X fez y gols;)
```

```python
# Abrindo os dicionários, listas e variáveis:
time = []
jogador = {}
gols = []
continuar = ''
avaliar = 0

#Criando a estrutura de repetição:
while True:
    
#Solicitando as informações do jogador:
    jogador['Nome'] = str(input('Digite o nome do jogador: '))
    jogador['Jogos'] = int(input(f'Digite o número de partidas que o {jogador["Nome"]} fez: '))
    for n in range(0, jogador['Jogos']):
        gols.append(int(input(f'Quantos gols o {jogador["Nome"]} no {n + 1}º jogo? ')))
    jogador['Gols'] = gols[:]
    jogador['Soma'] = sum(gols)
    time.append(jogador.copy())
    gols.clear()
    
#Verificador da estrutura de repetição com validação do termo informado pelo usuário:
    while True:
        continua = str(input('Deseja continuar [S/N]? '))
        if continua not in 'SsNn':
            print('Opção inválida! Digite S para sim ou N para não!')
        else:
            break
    if continua in 'Nn':
        break
    else:
        pass

#Retorno geral ao usuário:
print('-=' * 20)
print(f'{"SCOUT DO TIME":^40}')
print('-=' * 20)
print(f'{"Nº":<5}{"Jogador":<20}{"Total de gols":>15}')
for n in range(0, len(time)):
    print(f'{n:<5}{time[n]["Nome"]:<20}{time[n]["Soma"]:>10} gols')
print('-=' * 20)

#Avaliação individual do jogador:
while True:
    print()
    print('Deseja avaliar individualmente algum jogador?')
    avaliar = int(input('Digite o Nº do jogador ou informe um valor negativo para encerrar: '))
    if avaliar < 0:
        print('-=' *20)
        print('ENCERRANDO O PRAGRAMA, VOLTE SEMPRE!')
        break
    else:
        print()
        print('-=' *20)
        print(f'Scout individual do jogador {time[avaliar]["Nome"]:^11}')
        print('-=' *20)
        print(f'Jogos na temporada: {time[avaliar]["Jogos"]} jogos.')
        print('-'*40)
        print('Análise individual dos jogos:')
        for n in range(0, len(time[avaliar]["Gols"])):
            print(f'O jogador fez {time[avaliar]["Gols"][n]} gols na {n + 1} partida.')
        print(f'Totalizando {time[avaliar]["Soma"]} gols na temporada')
        print('-'*40)
        print('-=' *20)
```
