```python
#3) Crie um programa que vai gerar 5 números aleatórios e colocar em uma tupla e mostre:
#   - A listagem de números;
#   - O maior número;
#   - O menor número.

#Importando a função random
import random

#Inicializando as variárives de controle
maior = 0
menor = 0


#Gerando os 5 números aleatórios e vendo o maior e o menor
num_1 = random.randint(0,10)
menor = num_1
maior = num_1

num_2 = random.randint(0,10)
if num_2 > maior:
    maior = num_2
else:
    pass
if num_2 < menor:
    menor = num_2
else:
    pass

num_3 = random.randint(0,10)
if num_3 > maior:
    maior = num_3
else:
    pass
if num_3 < menor:
    menor = num_3
else:
    pass

num_4 = random.randint(0,10)
if num_4 > maior:
    maior = num_4
else:
    pass
if num_4 < menor:
    menor = num_4
else:
    pass

num_5 = random.randint(0,10)
if num_5 > maior:
    maior = num_5
else:
    pass
if num_5 < menor:
    menor = num_5
else:
    pass

#Colocando-os numa Tulpa
sorteados = (num_1, num_2, num_3, num_4, num_5)

#Mostrando a tulpa completa:
print('-=' * 20)
print(f'Os números sorteados foram: {sorteados}')

#Mostrando o número maior:
print('-=' * 20)
print(f'O maior número sorteado foi o {maior}')

#Mostrando o número menor:
print('-=' * 20)
print(f'O menor número sorteado foi o {menor}')

```
